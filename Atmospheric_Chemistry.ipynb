```python
#Simulates the reaction pathway energy surface for SO + nHO  HSO clusters (n=1-3)
#Predicts reaction rates at 60-100% relative humidity
#Maps quantum outputs to aerosol nucleation probabilities

def modelMolec(atomGeo):
  # Setup Geometry
  driver = PySCFDriver(
      atom=atomGeo,
      basis="sto3g",
      charge=0,
      spin=0,
      unit=DistanceUnit.ANGSTROM,
  )

  es_problem = driver.run()

  # Init mapper
  mapper = JordanWignerMapper()

  # Init Solver
  numpy_solver = NumPyMinimumEigensolver()

  # Init VQE Solver
  ansatz = UCCSD(
    es_problem.num_spatial_orbitals,
    es_problem.num_particles,
    mapper,
    initial_state=HartreeFock(
        es_problem.num_spatial_orbitals,
        es_problem.num_particles,
        mapper,
    ),
  )
  vqe_solver = VQE(Estimator(), ansatz, SLSQP())
  vqe_solver.initial_point = [0.0] * ansatz.num_parameters

  # Calculate result
  calc = GroundStateEigensolver(mapper, vqe_solver)
  res = calc.solve(es_problem)

  return res


# Solve each molecule
resSO3 = modelMolec("S 0.0000 0.0000 0.0000; O 0.0000 1.4175 0.0000; O 1.2276 -0.7088 0.0000; O -1.2276 -0.7088 0.0000")
resH2O = modelMolec("O 0.0000 0.0000 0.1173; H 0.0000 0.7572 -0.4692; H 0.0000 -0.7572 -0.4692")
resH2SO4 = modelMolec("S 0.0000 0.0000 0.1248; O 1.2510 -0.0349 0.8000; O -1.2510 0.0349 0.8000; O 0.0000 -1.2172 -0.8732; O 0.0000 1.2172 -0.8732; H -0.3548 1.9933 -0.4121; H 0.3548 -1.9933 -0.4121")

print(resSO3)
```
